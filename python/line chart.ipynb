```python
from vega_datasets import data
#Using the british monarch dataset that has continuous monarchy periods for over three centuries.
df = data.monarchs()
#Display the dataset
df
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 




fig, ax = plt.subplots()
#adjusting the dimensions of the figure
fig.set_size_inches(18.5, 10.5)

#Enabling X-ticks on the top axis and disabling on bottom
plt.rcParams['xtick.bottom'] = plt.rcParams['xtick.labelbottom'] = False
plt.rcParams['xtick.top'] = plt.rcParams['xtick.labeltop'] = True

#Using a qualitative color map for lines to represent the varying tenure periods of the monarchs.
cmap = plt.get_cmap("tab20b")

#Plotting each monarch's tenure as a horizontal line on a common Y(zero) to get a straight line
plt.hlines([0] * len(df), df['start'],df['end'], color=cmap(list(df.index)), linewidth=10)

#Using the start year as the ticks on the top X axis
plt.xticks(list(df['start']),  rotation=50)

#Printing the monarch name at the center and below each corresponding line
df.apply(lambda x : plt.text(x['start'] + (x['end'] -x['start'])/2,-0.3,x['name'], rotation=-50, fontstyle='italic') , axis = 1)

#Hiding all the spines and Y axis
ax.axes.get_yaxis().set_visible(False)
ax.axes.spines['right'].set_visible(False)
ax.axes.spines['top'].set_visible(False)
ax.axes.spines['left'].set_visible(False)
ax.axes.spines['bottom'].set_visible(False)

#Compressing the plot bottom for a nicer plot
plt.subplots_adjust(bottom=0.87)
plt.show()

```
